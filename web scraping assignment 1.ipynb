Q3. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating

```python
def scrape_odi_teams():
    icc_cricket=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
    icc_cricket
    soup=BeautifulSoup(icc_cricket.text,"html.parser")
    soup
    teams = []
    matches = []
    points = []
    ratings = []

    table = soup.find('table', class_='table')
    rows = table.tbody.find_all('tr')

    for row in rows:
        cols = row.find_all('td')
        team = cols[1].text.strip()
        match = cols[2].text.strip()
        point = cols[3].text.strip()
        rating = cols[4].text.strip()

        teams.append(team)
        matches.append(match)
        points.append(point)
        ratings.append(rating)

    data = {
        'Team': teams,
        'Matches': matches,
        'Points': points,
        'Rating': ratings
    }

    df = pd.DataFrame(data)
    return df.head(10)




if __name__ == '__main__':
    print("Top 10 ODI teams in men’s cricket:")
    print(scrape_odi_teams())

  

```

5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and
make data frame
i) Headline
ii) Time
iii) News Link

```python
url = "https://www.cnbc.com/world/?region=world"

response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

articles = soup.find_all("div", class_="Card-titleContainer")

headlines = []
times = []
links = []

for article in articles:
    
    headline = article.find("a").text.strip()
    headlines.append(headline)
    

    time_element = article.find("time")
    time = time_element.text.strip() if time_element else "N/A"
    times.append(time)
    
   
    link = article.find("a")["href"]
    links.append(link)


df = pd.DataFrame({"Headline": headlines, "Time": times, "News Link": links})


print(df)


```

```python

url=("https://www.dineout.co.in/delhi-restaurants/buffet-special")
dineout=requests.get(url)
soup=BeautifulSoup(dineout.content,'html.parser')
soup



```

```python
location1=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location1.append(i.text)
location1


 
```

```python
print(len('rest_name'),len('location1'))
```

```python
img1=[]
for i in soup.find_all("img",class_="no-img"):
    img1.append(i.get('data-src'))
img1
```

```python
rating=[]
for i in soup.find_all('span',class_="restnt-rating.rating-5"):
    rating.append(i.text)
rating
```
