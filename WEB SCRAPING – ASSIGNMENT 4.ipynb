```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up ChromeDriver path
chrome_driver_path = "path_to_chromedriver"

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode

# Start the ChromeDriver service
service = Service(chrome_driver_path)

# Start the WebDriver
driver = webdriver.Chrome(service=service, options=chrome_options)

# Navigate to the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

# Wait for the table to load
table_locator = (By.CLASS_NAME, "wikitable")
table = WebDriverWait(driver, 10).until(EC.presence_of_element_located(table_locator))

# Find all rows in the table (excluding the header row)
rows = table.find_elements(By.TAG_NAME, "tr")[1:]

videos = []
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    if len(cells) >= 6:  # Ensure that the row has at least 6 cells
        rank = cells[0].text.strip()
        name = cells[1].text.strip()
        artist = cells[2].text.strip()
        upload_date = cells[5].text.strip()
        views = cells[3].text.strip()

        video = {
            "Rank": rank,
            "Name": name,
            "Artist": artist,
            "Upload date": upload_date,
            "Views": views,
        }
        videos.append(video)
    else:
        print("Skipping row:", [cell.text.strip() for cell in cells])

# Print the scraped video details
for video in videos:
    print(video)

# Quit the WebDriver
driver.quit()

```

Q.4  Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

# Set up the Selenium web driver
driver=webdriver.Chrome(r"C:\Users\RB\Desktop\flip robo internship\New folder\chrome driver.exe")
  
driver.implicitly_wait(10)

# Open the website
url = "https://github.com/"
driver.get(url)

# Find and click on the "Trending" link
trending_link = driver.find_element(By.LINK_TEXT, "Trending")
trending_link.click()

# Wait for the trending repositories to load
repo_list = driver.find_elements(By.CSS_SELECTOR, ".Box-row")

# Scrape the details from each repository
for repo in repo_list:
    repo_title = repo.find_element(By.CSS_SELECTOR, "h1 a").text
    repo_description = repo.find_element(By.CSS_SELECTOR, "p").text

    # Open the repository page to get contributors and language
    repo_link = repo.find_element(By.CSS_SELECTOR, "h1 a")
    action = ActionChains(driver)
    action.key_down(Keys.CONTROL).click(repo_link).key_up(Keys.CONTROL).perform()

    # Switch to the new tab
    driver.switch_to.window(driver.window_handles[-1])

    # Wait for the repository page to load
    contributors_count_element = driver.find_element(By.CSS_SELECTOR, "a[data-tab-item='i4contributors-tab']")
    contributors_count = contributors_count_element.get_attribute("data-count")

    language_element = driver.find_element(By.CSS_SELECTOR, ".BorderGrid-cell:nth-child(2) span")
    language_used = language_element.text

    print("Repository Title:", repo_title)
    print("Repository Description:", repo_description)
    print("Contributors Count:", contributors_count)
    print("Language Used:", language_used)
    print("----------------------")

    # Close the repository tab
    driver.close()

    # Switch back to the main tab
    driver.switch_to.window(driver.window_handles[0])





```
